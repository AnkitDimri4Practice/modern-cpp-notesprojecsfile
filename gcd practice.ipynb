```python
# x=float(input("Enter the first value : "))
# y=float(input("Enter the Second value : "))
# sum = x+y
# print(sum)
# m=int(input("Enter the first value : "))
# n=int(input("Enter the Second value : "))
m=12
n=48
def gcd(m,n):
    fm = []
    for i in range(1,m+1):
        if (m%i)==0:
            fm.append(i)
    fn=[]
    for j in range(1,n+1):
        if (n%j)==0:
            fn.append(j)
    cf = []
    for f in fm:
        if f in fn:
            cf.append(f)
    return(cf[-1])
print("The GCD : ", gcd(m,n))
```
